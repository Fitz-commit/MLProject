```python
from my_datasets import *
import numpy as np
import matplotlib.pyplot as plt
# sklearn imports
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn import cluster
```

```python
data, _ = createData_2Gaussians()
# min-max scaling: determine scaling/normalization parameters
# (other scaling methods are possible, e.g. z-score using StandardScaler)
scaler = MinMaxScaler().fit(data)
data = scaler.transform(data)
1
#cluster the data with the specified number of cluster centers
kmeans = KMeans(n_clusters=2)
kmeans.fit(data) # sets the data in the kmeans object
plt.scatter(data[:, 0], data[:, 1], c = kmeans.labels_, cmap=plt.cm.coolwarm)
# plot the cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
marker='X', c="black")
plt.title('k-means with k = 2')
```

```python
# get data set
data, _ = createData_Moons()
# min-max scaling: determine scaling/normalization parameters
3
# (other scaling methods are possible, e.g. z-score using StandardScaler)
scaler = MinMaxScaler().fit(data)
data = scaler.transform(data)
#cluster the data with the specified number of cluster centers
kmeans = KMeans(n_clusters=2)
kmeans.fit_predict(data)
plt.scatter(data[:, 0], data[:, 1], c = kmeans.labels_, cmap=plt.cm.coolwarm)
# plot the cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
marker='X', c="black")
plt.title('k-means with k = 2')
```
